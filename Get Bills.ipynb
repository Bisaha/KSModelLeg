# Purpose
This notebook is meant to get bills from legiscan and put them into individual files so they can be used by the LID.

# Importing libraries

```python
from urllib.request import urlopen
import json
import base64
from bs4 import BeautifulSoup
import os
from time import sleep
import textract
```

# Setting variables

```python
variables = {
"session":"73",#2009,2010 session
"state":"ks",
"api" : os.environ.get("LEGISCAN_KEY")
}
```

# debug print statements switch

```python
print_on = True
```

## Make a request to the API

I'm asking for a masterlist of bills for the state of Kansas from the 2009, 2010 session. It is then decoded as utf-8

```python
r = urlopen("https://api.legiscan.com/?key={}&op=getMasterList&state={}&session_id={}".format(variables["api"],variables["state"],variables["session"])).read().decode('utf-8')
```

## Then turning the string into Json object

```python
json_obj = json.loads(r)
```

# then extracting the masterlist

```python
masterlist = json_obj.get("masterlist")
```

## Then looping through and making a list of bill IDs

```python
bill_id_list = []

for bill in masterlist.values():
    bill_id_list.append(bill.get("bill_id"))
bill_id_list
```

# Now to get the bill for each  bill id
for the get("texts") step, I'm looking at the first item in the text lists. This appears to be the origional, introduced bill. I'm working on the assumption that the ammendments wouldn't really tie into the "model" part of model legislation

```python
def get_doc_id(bill_id):
    searchId = urlopen('https://api.legiscan.com/?key={}&op=getBill&id={}'.format(variables["api"],bill_id)).read().decode()
    bill = json.loads(searchId)
    try:
        doc_id = bill.get("bill").get("texts")[0].get("doc_id")
    except:
        return bill_id
    return doc_id

doc_list = {}
no_doc_list = {}

for item in bill_id_list:
    if print_on is True:
        print("Getting doc id for {}".format(item))
    doc_id = get_doc_id(item)
    sleep(1)
    if doc_id == item:
        no_doc_list[item] = None
        if print_on is True:
            print("Bill {} has no doc".format(item))
    else:
        doc_list[item] = doc_id
        if print_on is True:
            print("Doc id is {}".format(doc_id))

    
doc_list


    
```
