# Purpose

This is notebook is designed to figure out how to get docs from legican and turn them into a readable format.

# Imports

```python
from urllib.request import urlopen
import os
import json
```

# practice variables

```python
variables = {
"doc_id":"1685967",
"key" : os.environ.get("LEGISCAN_KEY")
}
```

# API Call

```python
searchId = urlopen('https://api.legiscan.com/?key={}&op=getBillText&id={}'.format(variables["key"],variables["doc_id"])).read().decode()
```

# Get text

```python
js = json.loads(searchId)
text = js.get("text").get("doc")
text
```
